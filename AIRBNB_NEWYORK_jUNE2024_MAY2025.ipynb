```python
#Reading  the data & Creating Chunk For Calender
import pandas as pd
import zipfile
import os
import csv

# Extract ZIP
zip_path = '/content/calendar.zip'
extract_dir = '/content/calendar_extracted'
os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

#  Read and combine all .csv.gz files using chunks
calendar_parts = []
chunk_size = 10000

for file in os.listdir(extract_dir):
    if file.endswith('.csv.gz'):
        full_path = os.path.join(extract_dir, file)
        try:
            for chunk in pd.read_csv(full_path,
                                     compression='gzip',
                                     chunksize=chunk_size,
                                     on_bad_lines='skip',
                                     quoting=csv.QUOTE_MINIMAL,
                                     encoding='utf-8',
                                     engine='python'):
                calendar_parts.append(chunk)
        except Exception as e:
            print(f"Error reading {file}: {e}")

#  Concatenate all parts into one calendar_df
calendar_df = pd.concat(calendar_parts, ignore_index=True)
print(f"Loaded {len(calendar_df)} rows from all extracted calendar files.")
```

```python
# How does the price of Airbnb listings in New York City relate to their short- and long-term availability (30, 60, and 90 365 days)?


import plotly.express as px

# Make a copy to keep merged_df unchanged
availability30_60_90_365_price = merged_df.copy()

# Bin price into 10 quantile-based bins, drop duplicates if any
availability30_60_90_365_price['price_bin'] = pd.qcut(availability30_60_90_365_price['price'], q=10, duplicates='drop')

# Group by price bins and calculate average availability for 30, 60, 90, and 365 days
availability_by_price = availability30_60_90_365_price.groupby('price_bin')[['availability_30', 'availability_60', 'availability_90', 'availability_365']].mean().reset_index()

# Convert price_bin intervals to string for plotting
availability_by_price['price_bin'] = availability_by_price['price_bin'].astype(str)

# Melt to long format for plotly express
availability_long = availability_by_price.melt(
    id_vars='price_bin',
    var_name='Availability_Window',
    value_name='Average_Days_Available'
)

# Define the order of availability windows including 365 days
availability_order = ['availability_30', 'availability_60', 'availability_90', 'availability_365']

# Pivot table with ordered columns, renamed as requested
pivot_df_availability_30_60_90_365_price = availability_long.pivot(
    index='price_bin',
    columns='Availability_Window',
    values='Average_Days_Available'
)[availability_order]

# Plot heatmap with px.imshow
fig = px.imshow(
    pivot_df_availability_30_60_90_365_price,
    labels=dict(x="Availability Window (Days)", y="Price Range (Binned)", color="Avg. Availability"),
    color_continuous_scale='YlOrBr',
    aspect="auto",
    title="📊 How Does Price Affect Availability Across Short- and Long-Term Listing Availability?"
)

fig.update_layout(
    xaxis_title="Availability Period",
    yaxis_title="Price Range",
    autosize=True
)

fig.show()
```

```python
#Does the average listing price vary significantly based on the total number of nights available in the next 30 days?
import pandas as pd
from scipy.stats import f_oneway, kruskal

# Clean price column to numeric (remove '$', ',')
merged_df['price'] = merged_df['price'].astype(str).str.replace('[$,]', '', regex=True).astype(float)

# Define availability buckets based on availability_30
def availability_bucket(nights):
    if nights <= 10:
        return '0-10'
    elif 11 <= nights <= 20:
        return '11-20'
    elif 21 <= nights <= 30:
        return '21-30'
    else:
        return None

# Apply bucket function, drop rows where bucket is None or price is null
merged_df['availability_bucket'] = merged_df['availability_30'].apply(availability_bucket)
price_avail = merged_df.dropna(subset=['price', 'availability_bucket'])

# Group prices by availability bucket
prices_0_10 = price_avail[price_avail['availability_bucket'] == '0-10']['price']
prices_11_20 = price_avail[price_avail['availability_bucket'] == '11-20']['price']
prices_21_30 = price_avail[price_avail['availability_bucket'] == '21-30']['price']

# ANOVA test for mean price differences
anova_stat, anova_p = f_oneway(prices_0_10, prices_11_20, prices_21_30)

print(f"ANOVA F-statistic: {anova_stat:.3f}")
print(f"ANOVA p-value: {anova_p:.3f}")

if anova_p < 0.05:
    print("Reject H0: At least one group mean price is significantly different.")
else:
    print("Fail to reject H0: No significant difference in mean prices among groups.")

# Alternatively, Kruskal-Wallis (non-parametric)
kruskal_stat, kruskal_p = kruskal(prices_0_10, prices_11_20, prices_21_30)

print(f"Kruskal-Wallis statistic: {kruskal_stat:.3f}")
print(f"Kruskal-Wallis p-value: {kruskal_p:.3f}")

if kruskal_p < 0.05:
    print("Reject H0: At least one group distribution of prices differs.")
else:
    print("Fail to reject H0: No significant difference in price distributions among groups.")
```

```python
#How does minimum_nights and maximum_nights affect availability and price?
from scipy.stats import ttest_ind
import pandas as pd

# Clean price column (if not already done)
merged_df['price'] = (
    merged_df['price']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
)
merged_df['price'] = pd.to_numeric(merged_df['price'], errors='coerce')

# Filter out NaNs for relevant columns
df = merged_df[['minimum_nights', 'maximum_nights', 'availability_30', 'price']].dropna()

# Group 1: minimum_nights < 30
group_min_lt_30 = df[df['minimum_nights'] < 30]
group_min_ge_30 = df[df['minimum_nights'] >= 30]

# T-test: availability_30 vs minimum_nights
tstat1, pval1 = ttest_ind(group_min_lt_30['availability_30'], group_min_ge_30['availability_30'], equal_var=False)

# T-test: price vs minimum_nights
tstat2, pval2 = ttest_ind(group_min_lt_30['price'], group_min_ge_30['price'], equal_var=False)

# Group 2: maximum_nights < 365
group_max_lt_365 = df[df['maximum_nights'] < 365]
group_max_ge_365 = df[df['maximum_nights'] >= 365]

# T-test: availability_30 vs maximum_nights
tstat3, pval3 = ttest_ind(group_max_lt_365['availability_30'], group_max_ge_365['availability_30'], equal_var=False)

# Print results with interpretation
alpha = 0.05  # significance level

def interpret_result(t, p, var, condition):
    decision = "Reject" if p < alpha else "Fail to reject"
    conclusion = (
        f"T-test: {var} by {condition} => t = {t:.3f}, p = {p:.4f} → {decision} the null hypothesis.\n"
        f"There **{'is' if p < alpha else 'is no'}** statistically significant difference in {var} between the groups.\n"
    )
    return conclusion

print(interpret_result(tstat1, pval1, "availability_30", "minimum_nights <30 vs ≥30"))
print(interpret_result(tstat2, pval2, "price", "minimum_nights <30 vs ≥30"))
print(interpret_result(tstat3, pval3, "availability_30", "maximum_nights <365 vs ≥365"))
```

```python
# New York City's Local Law 18, short-term rentals are limited to a maximum of two guests at a time. This regulation applies regardless of the property's size or the number of bedrooms.
# Is the average price of listings accommodating 2 or fewer guests significantly lower than the average price of listings accommodating more than 2 guests?
# H0: The average price of listings accommodating 2 or fewer guests is greater than or equal to the average price of listings accommodating more than 2 guests.

# H1: The average price of listings accommodating 2 or fewer guests is less than the average price of listings accommodating more than 2 guests.


import numpy as np
from scipy.stats import ttest_ind

# Define two groups from merged dataframe
listings_accommodating_two_or_fewer_guests = merged_df[merged_df['accommodates'] <= 2]
listings_accommodating_more_than_two_guests = merged_df[merged_df['accommodates'] > 2]

# Extract prices from both groups
prices_two_or_fewer_guests = listings_accommodating_two_or_fewer_guests['price'].dropna()
prices_more_than_two_guests = listings_accommodating_more_than_two_guests['price'].dropna()

# Hypotheses:
# H0: mean price for listings accommodating ≤ 2 guests >= mean price for listings accommodating > 2 guests
# H1: mean price for listings accommodating ≤ 2 guests < mean price for listings accommodating > 2 guests

# Perform one-sided two-sample t-test (using equal_var=False for Welch's t-test)
t_statistic, p_value_two_sided = ttest_ind(prices_two_or_fewer_guests, prices_more_than_two_guests, equal_var=False)

# Since alternative is one-sided (less), adjust p-value accordingly
p_value_one_sided = p_value_two_sided / 2 if t_statistic < 0 else 1 - (p_value_two_sided / 2)

print(f"T-test statistic = {t_statistic:.3f}")
print(f"One-sided P-value = {p_value_one_sided:.4f}")

if p_value_one_sided < 0.05:
    print("Reject H0: Listings accommodating 2 or fewer guests have significantly lower average price.")
else:
    print("Fail to reject H0: No significant evidence that listings accommodating 2 or fewer guests have lower average price.")
```
